# Question 1
Please design a linear regression algorithm to obtain the optimal value of $w_0$ and $w_1$ for the given data in Data_HW1_1.csv by assuming the outputs and the data samples provided in the following satisfy $y=w_0+w_1\times x +\varepsilon$, where $\varepsilon$ is measurement noise.

*   Please provide the code, optimal values of $w_0$, $w_1$, training error, and testing error.

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, optimizers
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Obtain your data samples
file_direct = "C:\\Users\\LG\\Desktop\\github\\PURDUE-2023-FALL\\CNIT_483\HW_1\\"
data = pd.read_csv(file_direct+'Data_HW1_1.csv') # you may need to change the path
x_data = data['x'].values
y_data = data['y'].values

print(data)
# Generate training data (70% of the given data samples) and the testing data (30% of the given data samples). You can change to other percentage value as long as test_size <=0.3.
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3)
```

```python
# Gradient-descent solution
# Ref: 1) tf.keras.layers.Normalization: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Normalization; and 2) tf.keras.layers.Dense: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense
model = models.Sequential()
model.add(layers.Normalization(input_shape=(1,), axis=None)) # tighter, limiting the data into specific range
model.add(layers.Dense(1 ,activation = 'relu'))

model.summary()
```

```python
callback1 = tf.keras.callbacks.EarlyStopping(monitor='loss', patience = 50)
adam = optimizers.Adam(learning_rate=0.01)
adagd = optimizers.Adagrad(learning_rate=0.01)
SGD = optimizers.SGD(learning_rate=0.001)

model.compile(optimizer = adam,
              loss='mean_absolute_error',
              metrics=['mean_absolute_error'])

history = model.fit(x_train, y_train, epochs=2100,
                    validation_data=(x_test, y_test), callbacks=[callback1]) #, callbacks=[callback1])
```

# Q1 Train & Test Error

```python
# Visualize performance evaluation
plt.plot(history.history['mean_absolute_error'], label='Train_error')
plt.plot(history.history['val_mean_absolute_error'], label = 'Test_error')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.xlim([0, len(history.history['mean_absolute_error'])])
plt.ylim([4, 7])
plt.legend(loc='lower right')

results = model.evaluate(x_test, y_test, verbose = 0, callbacks=[callback1])
print(results)
# print('Train Error:', round(results[0],7), '\t', 'Test Error:',  round(results[1], 7))
```

# Question 2 
Please design a linear regression algorithm to obtain the optimal value of $w_0$, $w_1$, $w_2$, and $w_3$ for the given data in Data_HW1_2.csv by assuming the outputs and the data samples provided in the following satisfy $y=w_0+w_1\times x +w_2\times x^2+w_3\times x^3+\varepsilon$, where $\varepsilon$ is measurement noise.

*   Please provide the code, optimal values of $w_0$, $w_1$, $w_2$, $w_3$, training error, and testing error.

```python
data2 = pd.read_csv(file_direct+'\\Data_HW1_2.csv') # you may need to change the path

q2_x_data = data2['x'].values
x2_data = np.power(q2_x_data, 2)
x3_data = np.power(q2_x_data, 3)

x_final_data = np.stack([q2_x_data, x2_data, x3_data], axis = 1)

q2_y_data = data2['y'].values

print(x_final_data[:5])
# Generate training data (70% of the given data samples) and the testing data (30% of the given data samples). You can change to other percentage value as long as test_size <=0.3.
q2_x_train, q2_x_test, q2_y_train, q2_y_test = train_test_split(x_final_data, q2_y_data, test_size=0.3)
```

```python
# Gradient-descent solution
# Ref: 1) tf.keras.layers.Normalization: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Normalization; and 2) tf.keras.layers.Dense: https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense
model2 = models.Sequential()
model2.add(layers.Normalization(input_shape=(3,), axis=None)) # tighter, limiting the data into specific range
model2.add(layers.Dense(1 ,activation = 'relu'))

model2.summary()
```

```python
callback2 = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=33)
adam2 = optimizers.Adam(learning_rate=0.001)

model2.compile(optimizer = adam2,
              loss='mean_absolute_error',
              metrics=['mean_absolute_error'])

history2 = model2.fit(q2_x_train, q2_y_train, epochs=2100, validation_data=(q2_x_test, q2_y_test), callbacks=[callback2])
```

# Q2 Train & Test Error

```python
# Visualize performance evaluation
plt.plot(history2.history['mean_absolute_error'], label='Train_error')
plt.plot(history2.history['val_mean_absolute_error'], label = 'Test_error')
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.xlim([0, len(history2.history['mean_absolute_error'])])
plt.ylim([0, 265])
plt.legend(loc='lower right')
```

```python
results2 = model2.evaluate(q2_x_test, q2_y_test, verbose = 0)
print('Train Error:', round(results2[0],7), '\t', 'Test Error:',  round(results2[1], 7))
```
